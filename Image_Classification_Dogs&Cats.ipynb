```python
#Creating two empty list as data and labels
data = []
labels = []
```

```python
for category_idx, category in enumerate(Categories):
    for file in os.listdir(os.path.join(input_dir, category)):
        img_path = os.path.join(input_dir, category, file)
        img = imread(img_path) # image is as numpy array
        img = resize(img, (15,15))
        data.append(img.flatten())
        labels.append(category_idx)
        
```

```python
labels = np.asarray(labels)
data = np.asarray(data)
```

```python
labels.shape
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, shuffle=True, stratify=labels)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

sc.fit(X_train)
X_train_scaled = sc.transform(X_train)

X_test_scaled = sc.transform(X_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
```

```python
model_LR.fit(X_train, y_train)
```

```python
y_pred_LR = model_LR.predict(X_test)
y_pred_LR
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(accuracy_score(y_test, y_pred_LR))
print(confusion_matrix(y_test,y_pred_LR))
print(classification_report(y_test, y_pred_LR))
```

# Decision Tree

```python
from sklearn.model_selection import RandomizedSearchCV
#Define hyperparameters to be tuned and set range of values for each hyperparameter
hyperparameters = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 5, 10, 15, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}


#Create an instance of the RandomizedSearchCV model
random_search = RandomizedSearchCV(model_RF, hyperparameters, n_iter=10, cv=5, random_state=42)
```

```python
#Fit the RandomizedSearchCV model on the training data
random_search.fit(X_train, y_train)
```

```python
# Print the best hyperparameters selected by RandomizedSearchCV
print("Best hyperparameters: ", random_search.best_params_)

# Create a new instance of the RandomForestClassifier model using the best hyperparameters
mode_RF_best = RandomForestClassifier(n_estimators=random_search.best_params_['n_estimators'],
                                   max_depth=random_search.best_params_['max_depth'],
                                   min_samples_split=random_search.best_params_['min_samples_split'],
                                   min_samples_leaf=random_search.best_params_['min_samples_leaf'])

# Fit the new RandomForestClassifier model on the training data
mode_RF_best.fit(X_train, y_train)

# Generate predictions on the testing data using the new RandomForestClassifier model
y_pred_RF_best = mode_RF_best.predict(X_test)
```

```python
# Compute the accuracy score of the new RandomForestClassifier model

print(accuracy_score(y_test, y_pred_RF_best))
print(confusion_matrix(y_test,y_pred_RF_best))
print(classification_report(y_test, y_pred_RF_best))
```
